```python
## 라이브러리 불러오기
import matplotlib.pyplot as plt

## 한글폰트 적용하기
plt.rc('font', family='Malgun Gothic')
```

```python
sales = [1,5,7,3,7]
month = ['3월','4월','5월','6월','7월']

#제목
plt.title('월별 판매 실적')
# 범례, 색상
plt.plot(month, sales, label='단위:개수', color='g')
# 범례표시
plt.legend()
plt.show()
```

```python
sales = [1,5,7,3,7]
month = ['3월','4월','5월','6월','7월']

#제목
plt.title('월별 판매 실적')
# 범례, 색상, 차트의 모양 bar
plt.bar(month, sales, label='단위:개수', color='g')
# 범례표시
plt.legend()
plt.show()
```

### 2. 원형차트

```python
b_type = [25, 19, 37, 11]
b_name = ['A형','B형','O형','AB형']

# 원형차트, 범례, 비율
plt.pie(b_type, labels = b_name, autopct = '%1.1f%%')
plt.show()
```

### 3. Scatter 산점도

```python
kor =[80,20,50,20,10,50,60,30,40]
eng =[90,40,10,40,60,20,50,70,90]

# 색상, 투명도
plt.scatter(kor, eng, color='red', alpha=0.7)
# x,y축제목
plt.xlabel('국어 점수')
plt.ylabel('영어 점수')
plt.show()
```

```python
import random

height,weight,score = [],[],[]

for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    score.append(random.randint(1,1000))
    
#제목
plt.title('키와 몸무게와 체육점수의 관계')
# data, 색상, 투명도
plt.scatter(height,weight, s=score, c=score,cmap='rainbow', alpha=0.7)
plt.colorbar(label='체육 실기점수')
plt.show()
```

### 4. 상자 수염 그림 그리기

```python
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]

plt.title('오디션 프로그램 득표 현황')
plt.plot(singer, week1, label='첫째주', color='hotpink')
plt.plot(singer, week2, label='둘째주', color='royalblue')
plt.legend()
plt.show()
```
