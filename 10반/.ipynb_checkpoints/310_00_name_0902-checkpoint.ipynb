````python
#라이브러리 가져오기
import matplotlib.pyplot as plt

mon = ['3월','4월','5월','6월','7월']
sales = [1,5,7,3,7]

#한글폰트 적용하기
plt.rc('font', family='Malgun Gothic')
#차트 제목넣기
plt.title('월별 판매 실적')
#범례적용하기
plt.bar(mon, sales, color='b', label = '단위:개수')
plt.legend()
plt.show()
````

### plt.barh() : 막대그래프 - 가로형

````python
plt.title('월별 판매 실적')
plt.barh(mon,sales, color='g',label = '단위:개수')
plt.show()
````

### 히스토그램
### plt.hist()

````python
print(sales)
````

````python
plt.title('판매 실적 히스토그램')
plt.hist(sales)
plt.show()
````

### plt.pie() : 원그래프 

```
우리학교 100명의 혈액형을 조사한 결과가 다음과 같다.

A형 = 28명
B형 = 20명
O형 = 40명
AB형 = 12명 
```
#### 원그래프 각 항목에 라벨 붙이기 

```
plt.pie(리스트, labels = 라벨리스트)
```
#### 원그래프 항목에 퍼센트 표시

```
plt.pie(리스트, labels = 라벨리스트, autopct = '%1.1f%%')
```


````python
b_type = [28,20,40,12]
b_name = ['A형','B형','O형','AB형']

plt.title('혈액형 비율')
plt.pie(b_type, labels = b_name, autopct = '%1.1f%%')
plt.show()
````
