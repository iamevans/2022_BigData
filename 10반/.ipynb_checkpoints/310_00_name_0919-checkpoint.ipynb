````python
#차트 라이브러리 불러오기
import matplotlib.pyplot as plt

#한글폰트 적용하기
plt.rc('font', family='Malgun Gothic')
````

## 차트종류
```
꺽은선형 plt.plot
바형 plt.bar
가로 바형 plt.barh
원형차트 plt.pie
산점도 plt.scatter
상자 수업 그리기 plt.boxplot
```

````python
import random
````

````python
# 산점도 scatter
height, weight, score = [],[],[]
for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    score.append(random.randint(1,1000))
    
#제목 plt.title
plt.title('키와 몸무게와 체육점수의 관계')
plt.scatter(height, weight, s=score, c=score, cmap='rainbow', alpha=0.7)
plt.colorbar(label='체육실기점수')
plt.show()
````

````python
kor = [80, 20, 50, 20, 10, 50, 60, 30, 60] 
plt.boxplot(kor, labels=['국어점수'])
plt.show()
````

### 꺽은선형 그래프 겹쳐 그리기

````python
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]
plt.title('오디션 프로그램 특표 현황')
plt.plot(singer, week1, label='week1', color='hotpink')
plt.plot(singer, week2, label='week2', color='royalblue')
plt.legend()
plt.show()
````

````python
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]
plt.title('오디션 프로그램 특표 현황')

for i in range(len(week1)):
    week1[i] = -week1[i]
plt.rcParams['axes.unicode_minus'] = False #마이너스 기호표시
plt.barh(singer, week1, label='week1', color='hotpink')
plt.barh(singer, week2, label='week2', color='royalblue')
plt.legend()
plt.show()
````

````python
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]
plt.title('오디션 프로그램 특표 현황')
plt.bar(range(1,6), week1, label='week1', color='hotpink')
plt.bar(range(7,12), week2, label='week2', color='royalblue')
plt.xticks(range(1,12),['A','B','C','D','E',' ','A','B','C','D','E'])
plt.legend()
plt.show()
````

````python
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]
plt.title('오디션 프로그램 특표 현황')
plt.bar(range(1,14,3), week1, label='week1', color='hotpink')
plt.bar(range(2,15,3), week2, label='week2', color='royalblue')
plt.xticks(range(1,14,3), singer)
plt.legend()
plt.show()
````
