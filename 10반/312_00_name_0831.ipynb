````python
#라이브러리 가져오기 및 닉네임 설정
import matplotlib.pyplot as plt

#한글 폰트
plt.rc('font', family='Malgun Gothic')

mon = ['3월','4월','5월','6월','7월']
sales = [1,5,7,3,7]

#그래프의 제목
plt.title('월별 판매 실적')
#plot(옵션) x,y,color,label
plt.plot(mon,sales, color='g', label='단위:개수')
#범례표시 명령어
plt.legend()
plt.show()
````

### plt.bar() : 막대그래프 - 세로형
### plt.barh() : 막대그래프 - 가로형

````python
#라이브러리 가져오기 및 닉네임 설정
import matplotlib.pyplot as plt

#한글 폰트
plt.rc('font', family='Malgun Gothic')

mon = ['3월','4월','5월','6월','7월']
sales = [1,5,7,3,7]

#그래프의 제목
plt.title('월별 판매 실적')
#bar x,y,color,label
plt.bar(mon,sales, color='g', label='단위:개수')
#범례표시 명령어
plt.legend()
plt.show()
````
