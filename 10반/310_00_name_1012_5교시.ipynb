```python
#문제4 scatter chart

#라이브러리 불어오기
import matplotlib.pyplot as plt

#한글폰트 적용하기
plt.rc('font', family='Malgun Gothic')

#random
import random

height, weight = [],[]
for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    
#title
plt.title('키와 몸무게의 상관관계')
plt.scatter(height, weight, c=height, cmap='RdPu')
plt.show()
```

```python
#문제5 중복차트 그리기
#라이브러리 불어오기
import matplotlib.pyplot as plt

#한글폰트 적용하기
plt.rc('font', family='Malgun Gothic')

singer = [ 'A', 'B', 'C', 'D', 'E']
week1 = [42, 58, 19, 92, 84]
week2 = [53, 52, 48, 98, 73]
#title
plt.title('오디션 프로그램 득표 현황')
plt.plot(singer, week1, label = '첫째 주')
plt.plot(singer, week2, label = '둘째 주')
#범례
plt.legend()
plt.show()
```
