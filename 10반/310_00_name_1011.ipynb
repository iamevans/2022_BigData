```python
#문제1 꺽은선형 차트
#라이브러리 불러오기
import matplotlib.pyplot as plt
#한글폰트적용
plt.rc('font', family='Malgun Gothic')
#data
month = ['mar', 'apr', 'may', 'jun', 'jul']
sales = [1, 5, 7, 3, 7]
#title
plt.title('월별 판매 금액')
#plot chart
plt.plot(month, sales, label='단위 : 개수')
#legend
plt.legend(loc = 'lower right')
#xlabel,ylabel
plt.xlabel('월')
plt.ylabel('매출')
plt.show()
```

```python
#문제2 hist chart
#라이브러리 불러오기
import matplotlib.pyplot as plt
#한글폰트적용
plt.rc('font', family='Malgun Gothic')
#random 불러오기
import random
#blank list
a = []
for i in range(10):
    a.append(random.randint(1,10))
print(a)
plt.hist(a)
plt.show()
```

```python
#문제3 pie chart
#라이브러리 불러오기
import matplotlib.pyplot as plt
#한글폰트적용
plt.rc('font', family='Malgun Gothic')

#data
b_type = [25,  19, 37, 11]
b_name = ['AO형','BO형','OO형','AB형']

#pie chart
plt.pie(b_type, labels = b_name)
plt.show()
```
