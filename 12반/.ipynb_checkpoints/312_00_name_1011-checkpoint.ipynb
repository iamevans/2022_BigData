```python
#문제1 월별 판매 실적
#라이브러리 불러오기
import matplotlib.pyplot as plt

#한글 폰트 적용하기
plt.rc('font', family='Malgun Gothic')

month = ['mar','apr','may', 'jun', 'jul']
sales = [1,5,7,3,7]

#제목
plt.title('월별 판매 실적')
#꺽은선형 차트
plt.plot(month, sales, label = '단위 : 개수')

#범례
plt.legend(loc = 'lower right')

#x,y축제목
plt.xlabel('월')
plt.ylabel('매출')
plt.show()
```

```python
#문제2 히스토그램 그래프 그리기
#라이브러리 불러오기
import matplotlib.pyplot as plt

#한글 폰트 적용하기
plt.rc('font', family='Malgun Gothic')

#랜덤함수 불러오기
import random
a = []
for i in range(10):
    a.append(random.randint(1,10))
print(a)
plt.hist(a)
plt.show()
```

```python
#문제3 원형차트
#라이브러리 불러오기
import matplotlib.pyplot as plt

#한글 폰트 적용하기
plt.rc('font', family='Malgun Gothic')

#데이터
b_type = [25, 19, 37 ,11]
b_name = ['A형','B형','O형','AB형']

#원형차트
plt.pie(b_type, labels = b_name)

#원형차트 제목
plt.title('혈액형 원형 차트')

plt.show()
```

```python
#문제4 scatter chart
#라이브러리 불러오기
import matplotlib.pyplot as plt
#한글 폰트 적용하기
plt.rc('font', family='Malgun Gothic')
#random 불러오기
import random
#데이터 만들기
height, weight = [],[]
for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
#title
plt.title('키와 몸무게의 상관관계')
plt.scatter(height, weight, c=height, cmap='RdPu')
plt.show()
```

```python
#문제5 중복차트
#라이브러리 불러오기
import matplotlib.pyplot as plt
#한글 폰트 적용하기
plt.rc('font', family='Malgun Gothic')
#데이터 만들기
singer = ['A','B','C','D','E']
week1 = [42, 58, 19, 92, 84]
week2 = [53, 52, 48, 98, 73]

#title
plt.title('오디션 프로그램 득표 현황')
plt.plot(singer, week1, label = 'week1')
plt.plot(singer, week2, label = 'week2')
plt.legend()
plt.show()
```
