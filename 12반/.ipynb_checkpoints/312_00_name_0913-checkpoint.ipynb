```python
#라이브러리 가져오기
import matplotlib.pyplot as plt

#한글 폰트 적용하기
plt.rc('font', family='Malgun Gothic')
```

```python
sales = [1, 5, 7, 3 , 7]
month = ['3월', '4월', '5월', '6월', '7월']

#꺽은선형 
plt.plot(month, sales)
plt.show()
```

```python
#bar형 그려보기
plt.bar(month, sales)
plt.show()
```

```python
#bar형, 범례 단위:개수
#차트 제목 '월별 판매 실적'
plt.title('월별 판매 실적')
plt.bar(month,sales, label='단위:개수', color='g')
plt.legend()
plt.show()
```

```python
#원형차트
b_type = [25,19,37,11]
b_name = ['A형','B형','O형','AB형']

#원차트 명령어
plt.pie(b_type, labels = b_name, autopct = '%1.1f%%')
plt.show()
```
