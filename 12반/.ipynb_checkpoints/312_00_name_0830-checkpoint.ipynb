### 선 그래프 그리기

````python
import matplotlib.pyplot as plt
plt.title('월별 판매 현황')
sales = [1,5,7,3,7]

plt.plot(sales)
plt.show()
````

### x축 추가하기
### plot 꺽은선형 그래프
```
plt.plot(x축, y축)
plt.show()
```

````python
mon = ['3월','4월','5월','6월','7월']
sales = [1,5,7,3,7]

plt.title('월별 판매 현황')
plt.plot(mon, sales)
plt.show()
````

### 그래프 색상 변경


```
# 그래프 그리는 명령어에 color 옵션 추가
plt.plot(x축, y축, color = 'r')
```


````python
mon = ['3월','4월','5월','6월','7월']
sales = [1,5,7,3,7]

plt.title('월별 판매 현황')
plt.plot(mon, sales, 'g') # color생략가능
plt.show()
````

### 범례 나타내기 

```
plt.plot(x축, y축, label = '범례 내용')
plt.legend()      # 범례를 화면에 나타내기
plt.show()
```

````python
mon = ['3월','4월','5월','6월','7월']
sales = [1,5,7,3,7]

plt.title('월별 판매 현황')
plt.plot(mon, sales, color ='g', label ='단위:개수' ) 
plt.legend()
plt.show()
````

## 그래프 그리는 포맷
```
import matplotlib.pyplot as plt

plt.title('그래프 제목')
plt.plot(x축 데이터, y축 데이터)
plt.show()
```
그래프 명령어 plt.plot()에 사용가능한 옵션

```
color = '색상이름'             
label = '범례내용'
```
기타 사용가능한 명령어
```
plt.figure(dpi=150)     # 그래프 해상도 설정 
plt.legend()            # 범례 나타내기
```

````python
mon = ['3월','4월','5월','6월','7월']
sales = [1,5,7,3,7]

plt.figure(dpi=100)
plt.title('월별 판매 현황')
plt.plot(mon, sales, color ='g', label ='단위:개수' ) 
plt.legend()
plt.show()
````
