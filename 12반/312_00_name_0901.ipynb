```python
#라이브러리 가져오기
import matplotlib.pyplot as plt

#한글폰트
plt.rc('font', family='Malgun Gothic')
```

```python
sales = [1,5,7,3,7]

#차트제목
plt.title('월별 판매 현황')
#bar 차트, x,y,color,label
plt.bar(range(3,8), sales, color='b', label = '단위:개수')
plt.legend()
plt.show()
```

### plt.barh() : 막대그래프 - 가로형

```python
#차트제목
plt.title('세로막대형 월별 판매 현황')
#bar 차트, x,y,color,label
plt.barh(range(3,8), sales, color='b', label = '단위:개수')
plt.legend()
plt.show()
```

### plt.hist() :히스토그램

```python
print(sales)
```

```python
plt.rc('font', family='Malgun Gothic')

plt.title('판매 실적 히스토그램')
plt.hist(sales)
plt.show()
```

```python
#random 함수를 이용한 히스토그램

import random

a = []
for i in range(10):
    a.append(random.randint(1,10))
print(a)
plt.hist(a)
plt.show()
```
