````python
# 라이브러리 불러오기
import matplotlib.pyplot as plt

#한글폰트 적용하기
plt.rc('font', family='Malgun Gothic')
````

````python
#x,y값
mon = range(3,8)
sales =  [1,5,7,3,7]

#제목
plt.title('월별 판매 실적')
#차트 종류
#꺽은선형 plt.plot
#plt.plot(mon, sales)
#bar chart plt.bar
plt.bar(mon, sales, color='g', label='단위:개수')
#범례명령어 plt.legend()
plt.legend()
#보기명령어
plt.show()
````

````python
#혈액형 인원수
b_type = [28,20,40,12]
#혈액형 
b_name = ['A형','B형','O형','AB형']
#제목
plt.title('혈액형 비율')
#파이차트 plt.pie()
plt.pie(b_type, labels = b_name, autopct='%1.1f%%')
plt.show()
````

````python
kor =[80, 20, 50, 20, 10, 50, 60, 30, 60]
eng =[90, 40, 60, 40, 10, 30, 50, 70, 90]

#제목
plt.title('국어점수와 영어점수와의 상관관계')
plt.scatter(kor,eng, color='indigo')
#x축
plt.xlabel('국어점수')
#y축
plt.ylabel('영어점수')
plt.show()
````
