```python
#라이브러리 가져오기
import matplotlib.pyplot as plt

#한글 폰트 적용하기
plt.rc('font', family='Malgun Gothic')
```

```python
sales = [1, 5, 7, 3 , 7]
month = ['3월', '4월', '5월', '6월', '7월']

#꺽은선형 
plt.plot(month, sales)
plt.show()
```

```python
#bar형 그려보기
plt.bar(month, sales)
plt.show()
```

```python
#bar형, 범례 단위:개수
#차트 제목 '월별 판매 실적'
plt.title('월별 판매 실적')
plt.bar(month,sales, label='단위:개수', color='g')
plt.legend()
plt.show()
```

```python
#원형차트
b_type = [25,19,37,11]
b_name = ['A형','B형','O형','AB형']

#원차트 명령어
plt.pie(b_type, labels = b_name, autopct = '%1.1f%%')
plt.show()
```

```python
#산점도 그리기

kor =[80,20,50,20,10,50,60,30,40]
eng =[90,40,10,40,60,20,50,70,90]

#산점도, color 책상, alpha 투명도
plt.scatter(kor, eng, color='r', alpha=0.7)
#x축,y축 제목
plt.xlabel('국어점수')
plt.ylabel('영어점수')
plt.show()
```

```python
#키와 몸무게의 상관관계

import random
height = []
weight = []
for i in range(100) :
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))

#제목 '키와 몸무게의 상관관계
plt.title('키와 몸무게의 상관관계')
plt.scatter(height,weight, c= height, cmap='RdPu')
plt.xlabel('키')
plt.ylabel('몸무게')
plt.colorbar(label='키')
plt.show()
```

```python
#키와 몸무게 체육실기점수의 관계
height, weight, score = [],[],[]

for i in range(100) :
    height.append(random.randint(100, 200))
    weight.append(random.randint(20, 100))
    score.append(random.randint(1, 1000))
plt.title('키와 몸무게와 실기점수의 관계')
plt.scatter(height, weight, s = score, c = score ,cmap ='rainbow', alpha=0.7)
plt.colorbar(label='체육 실기점수')
plt.show()
```
